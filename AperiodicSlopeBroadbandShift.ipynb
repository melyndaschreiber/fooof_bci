```python
eeg_Names = ["F3", "F1", "Fz", "F2", "F4", "FFC5h", "FFC3h", "FFC1h", "FFC2h", "FFC4h", 
             "FFC6h", "FC5", "FC3", "FC1", "FCz", "FC2", "FC4", "FC6", "FTT7h", "FCC5h",
            "FCC3h", "FCC1h", "FCC2h", "FCC4h", "FCC6h", "FTT8h", "C5", "C3", "C1", "Cz",
            "C2", "C4", "C6", "TTP7h", "CCP5h", "CCP3h", "CCP1h", "CCP2h", "CCP4h", "CCP6h", 
            "TTP8h", "CP5", "CP3", "CP1", "CPz", "CP2", "CP4", "CP6", "CPP5h", "CPP3h", 
            "CPP1h", "CPP2h", "CPP4h", "CPP6h", "P3", "P1", "Pz", "P2", "P4", "PPO1h", 
            "PPO2h"]
```

```python
# Create a fucntion that takes paramers data dictionary and data labels to output a dataframe
def function(data_dictionary, data_labels, num_subjects_start, eeg_Names):
    # Create empty array to append information
    Subject = []
    Events = []
    Slope = []
    Offset = []
    Flat_Alpha, Flat_Beta = [], []
    Original_Alpha, Original_Beta = [], []
        
    slope_Names, offset_Names, flat_alpha_Names, flat_beta_Names, original_alpha_Names, original_beta_Names = [], [], [], [], [], []
    alpha = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O']
    for i in range(len(eeg_Names)): 
        slope_Names.append(eeg_Names[i] + '_slope')
        offset_Names.append(eeg_Names[i] + '_offset')
        flat_alpha_Names.append(eeg_Names[i] + '_flat_alpha')
        flat_beta_Names.append(eeg_Names[i] + '_flat_beta')
        original_alpha_Names.append(eeg_Names[i] + '_original_alpha')
        original_beta_Names.append(eeg_Names[i] + '_original_beta')

    # Loop through the data frames
    for i in range(len(data_labels)):
        for j in range(len(data_labels[i])): 
            electrode_slope, electrode_shift = [], []
            for k in range(len(data_labels[i][j])):
                Subject.append(alpha[i+num_subjects_start-1])
                this_event = data_labels[i][j][k]
                if this_event == 0: this_event = 'Rest'
                if this_event == 1: this_event = 'Elbow_Flex'
                if this_event == 2: this_event = 'Elbow_Extend'
                if this_event == 3: this_event = 'Supination'
                if this_event == 4: this_event = 'Pronation'
                if this_event == 5: this_event = 'Hand_Close'
                if this_event == 6: this_event = 'Hand_Open'
                Events.append(this_event)
                
                all_electrode_slopes = []
                all_electrode_offsets = []
                alpha_flat = []
                beta_flat = []
                alpha_original = []
                beta_original = []
                for l in range(len(data_dictionary[i][j][0][k])):
                    electrode_slopes = data_dictionary[i][j][0][k][l]['Slope']
                    electrode_offsets = data_dictionary[i][j][0][k][l]['Offset']
                    all_electrode_slopes.append(electrode_slopes)
                    all_electrode_offsets.append(electrode_offsets)
                    
                    alpha_flat_mean = np.mean(data_dictionary[i][j][0][k][l]['flat_Spectrum'][7:11])
                    beta_flat_mean = np.mean(data_dictionary[i][j][0][k][l]['flat_Spectrum'][15:23])
                    alpha_flat.append(alpha_flat_mean)
                    beta_flat.append(beta_flat_mean)
                    
                    alpha_original_mean = np.mean(data_dictionary[i][j][0][k][l]['Foofed'][7:11])
                    beta_original_mean = np.mean(data_dictionary[i][j][0][k][l]['Foofed'][15:23])
                    alpha_original.append(alpha_original_mean)
                    beta_original.append(beta_original_mean)
                    
                Slope.append(all_electrode_offsets)
                Offset.append(all_electrode_slopes)
                Flat_Alpha.append(alpha_flat)
                Flat_Beta.append(beta_flat)
                Original_Alpha.append(alpha_original)
                Original_Beta.append(beta_original)
                
                #event_slopes.append(all_electrode_slopes)
                #event_offsets.append(all_electrode_offsets)

    # Create two dictionaries
    conditions = {'Subject': Subject, 'Events': Events}
    #results = {'Slope': Slope, 'Offset': Offset}

    # create two dataframes
    Conditions = pd.DataFrame.from_dict(conditions) 
    slope_df = pd.DataFrame(Slope)
    offset_df = pd.DataFrame(Offset)
    original_beta_df = pd.DataFrame(Original_Beta)
    original_alpha_df = pd.DataFrame(Original_Alpha)
    flat_alpha_df = pd.DataFrame(Flat_Alpha)
    flat_beta_df = pd.DataFrame(Flat_Beta)
    
    condition_header = ['Subject', 'Event']
    header = condition_header + offset_Names + slope_Names + flat_alpha_Names + flat_beta_Names + original_alpha_Names + original_beta_Names

    # Bind two dataframes
    data_frame = pd.concat([Conditions, slope_df, offset_df, flat_alpha_df, flat_beta_df, original_alpha_df, original_beta_df], axis = 1, sort = False)
    data_frame.columns = header

    return data_frame

data_int_1 = function(data_results_1_2, data_labels_1_2, 1, eeg_Names)
data_int_2 = function(data_results_3_4, data_labels_3_4, 3, eeg_Names)
data_int_3 = function(data_results_5_6, data_labels_5_6, 5, eeg_Names)
data_int_4 = function(data_results_7_8, data_labels_7_8, 7, eeg_Names)
data_int_5 = function(data_results_9_10, data_labels_9_10, 9, eeg_Names)
data_int_6 = function(data_results_11_12, data_labels_11_12, 11, eeg_Names)
data_int_7 = function(data_results_13_15, data_labels_13_15, 13, eeg_Names)

combined_results = pd.concat([data_int_1, data_int_2, data_int_3, data_int_4, data_int_5, data_int_6, data_int_7], 
                             axis = 0, sort = False)

# data_int_1 = function(data_results_128_1_2, data_labels_128_1_2, 1, eeg_Names)
# data_int_2 = function(data_results_128_3_4, data_labels_128_3_4, 1, eeg_Names)
# combined_results = pd.concat([data_int_1, data_int_2, 1, eeg_Names])
```

```python
# Create a fucntion that takes paramers data dictionary and data labels to output a dataframe
def encoded_labels(data_dictionary, data_labels, num_subjects_start, eeg_Names):
    # Create empty array to append information
    Subject = []
    Events = []
    Slope = []
    Offset = []
    Flat_Alpha, Flat_Beta = [], []
    Original_Alpha, Original_Beta = [], []
        
    slope_Names, offset_Names, flat_alpha_Names, flat_beta_Names, original_alpha_Names, original_beta_Names = [], [], [], [], [], []
    alpha = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'M', 'N', 'O']
    for i in range(len(eeg_Names)): 
        slope_Names.append(eeg_Names[i] + '_slope')
        offset_Names.append(eeg_Names[i] + '_offset')
        flat_alpha_Names.append(eeg_Names[i] + '_flat_alpha')
        flat_beta_Names.append(eeg_Names[i] + '_flat_beta')
        original_alpha_Names.append(eeg_Names[i] + '_original_alpha')
        original_beta_Names.append(eeg_Names[i] + '_original_beta')

    # Loop through the data frames
    for i in range(len(data_labels)):
        for j in range(len(data_labels[i])): 
            electrode_slope, electrode_shift = [], []
            for k in range(len(data_labels[i][j])):
                Subject.append(alpha[i+num_subjects_start-1])
                this_event = data_labels[i][j][k]
                Events.append(this_event)
                
                all_electrode_slopes = []
                all_electrode_offsets = []
                alpha_flat = []
                beta_flat = []
                alpha_original = []
                beta_original = []
                for l in range(len(data_dictionary[i][j][0][k])):
                    electrode_slopes = data_dictionary[i][j][0][k][l]['Slope']
                    electrode_offsets = data_dictionary[i][j][0][k][l]['Offset']
                    all_electrode_slopes.append(electrode_slopes)
                    all_electrode_offsets.append(electrode_offsets)
                    
                    alpha_flat_mean = np.max(data_dictionary[i][j][0][k][l]['flat_Spectrum'][7:11])
                    beta_flat_mean = np.max(data_dictionary[i][j][0][k][l]['flat_Spectrum'][15:23])
                    alpha_flat.append(alpha_flat_mean)
                    beta_flat.append(beta_flat_mean)
                    
                    alpha_original_mean = np.max(data_dictionary[i][j][0][k][l]['Foofed'][7:11])
                    beta_original_mean = np.max(data_dictionary[i][j][0][k][l]['Foofed'][15:23])
                    alpha_original.append(alpha_original_mean)
                    beta_original.append(beta_original_mean)
                    
                Slope.append(all_electrode_offsets)
                Offset.append(all_electrode_slopes)
                Flat_Alpha.append(alpha_flat)
                Flat_Beta.append(beta_flat)
                Original_Alpha.append(alpha_original)
                Original_Beta.append(beta_original)
                
                #event_slopes.append(all_electrode_slopes)
                #event_offsets.append(all_electrode_offsets)

    # Create two dictionaries
    conditions = {'Subject': Subject, 'Events': Events}
    #results = {'Slope': Slope, 'Offset': Offset}

    # create two dataframes
    Conditions = pd.DataFrame.from_dict(conditions) 
    slope_df = pd.DataFrame(Slope)
    offset_df = pd.DataFrame(Offset)
    original_beta_df = pd.DataFrame(Original_Beta)
    original_alpha_df = pd.DataFrame(Original_Alpha)
    flat_alpha_df = pd.DataFrame(Flat_Alpha)
    flat_beta_df = pd.DataFrame(Flat_Beta)
    
    condition_header = ['Subject', 'Event']
    header = condition_header + offset_Names + slope_Names + flat_alpha_Names + flat_beta_Names + original_alpha_Names + original_beta_Names

    # Bind two dataframes
    data_frame = pd.concat([Conditions, slope_df, offset_df, flat_alpha_df, flat_beta_df, original_alpha_df, original_beta_df], axis = 1, sort = False)
    data_frame.columns = header

    return data_frame

data_int_1 = encoded_labels(data_results_1_2, data_labels_1_2, 1, eeg_Names)
data_int_2 = encoded_labels(data_results_3_4, data_labels_3_4, 3, eeg_Names)
data_int_3 = encoded_labels(data_results_5_6, data_labels_5_6, 5, eeg_Names)
data_int_4 = encoded_labels(data_results_7_8, data_labels_7_8, 7, eeg_Names)
data_int_5 = encoded_labels(data_results_9_10, data_labels_9_10, 9, eeg_Names)
data_int_6 = encoded_labels(data_results_11_12, data_labels_11_12, 11, eeg_Names)
data_int_7 = encoded_labels(data_results_13_15, data_labels_13_15, 13, eeg_Names)

combined_encoded_results = pd.concat([data_int_1, data_int_2, data_int_3, data_int_4, data_int_5, data_int_6, data_int_7], 
                             axis = 0, sort = False)
```

```python
# Plot the peak frequencies (x-axis) and amplitude (y-axis) of the guassians
# Get subject1 data
# Select the Elbow_Flex and Rest Events
# [Subject][Trial][0][Events][Electrodes]['Parameters']
#len(data_results_1_2[0][0][0])

# Make sure you add appropiate labels
def create_peak_data(data, labels, num_sub_start):
    data_results_1_2, data_labels_1_2 = data, labels
    this_Subject, this_Trial, this_Event, this_Elect, this_Peak, this_Amp, this_BW  = [], [], [], [], [], [], []
    for i in range(len(data_results_1_2)): # subject
        for j in range(len(data_results_1_2[i])): # trials
            for k in range(len(data_results_1_2[i][j][0])): # Random 0
                for l in range(len(data_results_1_2[i][j][0][k])): # Event
                    for m in range(len(data_results_1_2[i][j][0][k][l])): # Electrode
                        for n in range(len(data_results_1_2[i][j][0][k][l]['GaussianParams'])): # Various peaks
                            this_Subject.append(i + num_sub_start)
                            this_Trial.append(j)
                            this_Event.append(data_labels_1_2[i][j][k])
                            this_Elect.append(eeg_Names[l])
                            this_Peak.append(data_results_1_2[i][j][0][k][l]['GaussianParams'][n,0])
                            this_Amp.append(data_results_1_2[i][j][0][k][l]['GaussianParams'][n,1])
                            this_BW.append(data_results_1_2[i][j][0][k][l]['GaussianParams'][n,2])

    this_peak_data = {'Subject': this_Subject, 'Trial': this_Trial, 'Events': this_Event, 'Electrode': this_Elect,
                'Peak': this_Peak, 'Amp': this_Amp, 'BW': this_BW}

    peak_data = pd.DataFrame(this_peak_data)
    return peak_data
```
